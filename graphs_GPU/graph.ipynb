```python
import re
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
log_serial_time = [] 
log_serial_particle = []

f = open("serial_compare.txt", "r")
for line in f:
    spl = [float(x) for x in re.findall(r"[-+]?\d*\.\d+|\d+", line)]
    log_serial_particle.append(np.log(spl[0]) )
    log_serial_time.append(np.log(spl[1]) )
    

log_mpi_time = []
log_mpi_particle = [] 
f = open("mpi_compare.txt", "r")
for line in f:

    spl = [float(x) for x in re.findall(r"[-+]?\d*\.\d+|\d+", line)]
    log_mpi_time.append(np.log(spl[1]) )
    log_mpi_particle.append(np.log(spl[0]))
    
log_openmp_time = []
log_openmp_particle = [] 
f = open("openmp_compare.txt", "r")
for line in f:

    spl = [float(x) for x in re.findall(r"[-+]?\d*\.\d+|\d+", line)]
    log_openmp_time.append(np.log(spl[1]) )
    log_openmp_particle.append(np.log(spl[0]))
    
log_gpu_time = []
log_gpu_particle = [] 
f = open("gpu_compare.txt", "r")
for line in f:

    spl = [float(x) for x in re.findall(r"[-+]?\d*\.\d+|\d+", line)]
    log_gpu_time.append(np.log(spl[1]) )
    log_gpu_particle.append(np.log(spl[0]))
```

```python
fig=plt.figure()
ax1=fig.add_subplot(111, label="1")
ax2=fig.add_subplot(111, label="2", frame_on=False)
ax3=fig.add_subplot(111, label="3", frame_on=False)
ax4=fig.add_subplot(111, label="4", frame_on=False)

line1, = ax1.plot(log_serial_particle, log_serial_time, "-o", color="green", label = "Serial")

line2, = ax2.plot(log_mpi_particle, log_mpi_time, "-o", color = "purple", label = "MPI 32 Threads")

line3, = ax3.plot(log_openmp_particle, log_openmp_time, "-o", color = "red", label = "OpenMP 32 Threads")

line4, = ax4.plot(log_gpu_particle, log_gpu_time, "-o", color = "blue", label = "GPU")

ax2.set_xlabel("Log of Particles", color = "black")
ax2.set_ylabel('Log of Time', color="black")        
#ax2.yaxis.set_label_position('right') 
#ax2.tick_params(axis='y', colors="blue")

ax1.set_ylim(-3.0,3.0)
ax2.set_ylim(-3.0, 3.0)
ax3.set_ylim(-3.0,3.0)
ax4.set_ylim(-3.0, 3.0)

# ax1.legend(loc = (8.7, 2.0))
# ax2.legend(loc = (9.7, 2.0))
# ax3.legend()
# ax4.legend()
ax1.legend((line1, line2, line3, line4), ('Serial', 'MPI 32 Th', 'OMP 32 Th', 'GPU'), loc = 2)

plt.title("Plot of all Methods on Particles")
plt.show()
```
